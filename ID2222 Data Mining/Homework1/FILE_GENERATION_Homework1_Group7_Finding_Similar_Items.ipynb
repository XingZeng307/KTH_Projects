```python
import fileinput
import datetime
import itertools
import time

def Time_Stamp():
    date_time_now = datetime.datetime.now()
    
    D = str(date_time_now.day)
    M = str(date_time_now.month)
    Y = str(date_time_now.year)

    h = str(date_time_now.hour)
    m = str(date_time_now.minute)
    s = str(date_time_now.second)
    
    date_array = [D, M, Y, h, m, s]
    
    return date_array
    
def FileNameUnique(prefix = "data_temp/Movies_", suffix = '.txt'):
    file_name = prefix

    date_array = Time_Stamp()
    
    for idx, i in enumerate(date_array):
        if idx == 2:
            file_name += i + '_'
        elif idx == 5:
            file_name += i + suffix
        else:
            file_name += i + '.'

    return file_name

def Read_Movie_Data(file_path, start_line, n_lines, output_to='file'):
    movie_data = ""
    total_lines_read = 0
    
    file_read = open(file_path, "r", encoding="latin-1")
    total_lines_read = 0
    end_line = start_line + n_lines

    while total_lines_read >= start_line and total_lines_read < end_line: 
        # Get next line from file 
        line = file_read.readline() 

        if line != "\n":
            movie_data += line.strip() + " "
            total_lines_read += 1

    file_read.close()
    movie_data = movie_data
    
    if output_to == 'file':
        # to avoid overwriting an existing file, if you want to overwrite, just basically use output.txt
        file_name = FileNameUnique() 
        with open(file_name, "w") as output:
            output.write(movie_data)
    elif output_to == 'screen':
        #if you just want to show it here
        print(movie_data)
        
        
def Read_Movie_Data_Multiple_Files(file_path, start_line = 0, n_lines = 1000, n_files = 10, output_to = 'file'):
    movie_data = ['']
    total_lines_read = 0
    
    file_read = open(file_path, "r", encoding="latin-1")
    total_lines_read = 0
    end_line = start_line + n_lines * n_files
    
    file_processing = 0
    next_file_at = start_line + n_lines

    while total_lines_read >= start_line and total_lines_read < end_line: 
        # Get next line from the file 
        line = file_read.readline() 

        if line != "\n":
            movie_data[file_processing] += line.strip() + " "
            
        total_lines_read += 1

        if total_lines_read == next_file_at:
            print("next_file_at: " + str(next_file_at))
            
            if output_to == 'file':
            # to avoid overwriting an existing file... if you want to overwrite, just basically use output.txt
                file_name = FileNameUnique() 
                with open(file_name, "w") as output:
                    output.write(movie_data[file_processing]) 
                
                time.sleep(1)
            elif output_to == 'screen':
                #if you just want to show it here, uncomment to above part
                print(movie_data[file_processing])
                      
            file_processing += 1
            next_file_at += n_lines
            movie_data.append('')
            
    file_read.close()
    
    
```
