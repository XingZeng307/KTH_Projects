```python
# generate candidate itemset from last frequent itemset Lk_minus-1
# paras: Transaction: All transactions; Lk_minus_1: the last frequent itemset L(k-1)
# return: candidate k-itemset: dictionary {("itemset":support),("itemset1":support1)}
def generate_Ck(Transaction, Lk_minus_1):
    L_list = [i for i in Lk_minus_1.keys()] # eg k-1=2: [('39', '825'), ('704', '825'), ('39', '704'), ('227', '390')]
    Ck = {}
    for transaction in Transaction:
        L_list_rep = [] #('39', '825'), ('704', '825')  avoid computing the same itemset multiple times in one transcation
        for index_i in range(len(L_list)): # first for loop last frequent itemset L(k-1) # fisrt element
            if all(ele in transaction for ele in L_list[index_i]):
                for index_j in range(len(L_list)): # second for loop last frequent itemset L(k-1) to generate new subset
                    combine_tuple = L_list[index_j]
                    for combine_ele in combine_tuple: # second one element
                        if combine_ele not in L_list[index_i] and combine_ele in transaction:
                            new_tuple = L_list[index_i] + (combine_ele,) # generate subset if in transaction and no repetation in tuple
                            # order tuple
                            temp_list_t1 = [int(i) for i in new_tuple]
                            temp_list_t1.sort()
                            temp_list_t2 = [str(i) for i in temp_list_t1 ]
                            new_tuple_ordered = tuple(temp_list_t2)
                            if new_tuple_ordered not in L_list_rep: # no repeatitive calculation in one row(transcation)
                                        L_list_rep.append(new_tuple_ordered)
                                        if new_tuple_ordered not in Ck.keys():
                                            Ck[new_tuple_ordered] = 1
                                        else:
                                            Ck[new_tuple_ordered] += 1
    return Ck
                    
        
    
```

```python
# generate frequent k-itemset whose support > threshold support
def generate_Lk(Ck,support):
    Lk = {}    # e.g. return {("doubleton",support)}
    for key, value in Ck.items():
        if value > support-1:
            Lk[key] = value
    return Lk
```

```python
# the main function
# 1st step: read data
datContent = [i.strip().split() for i in open("./T10I4D100K.dat").readlines()]
# 2nd step: set support and confidence
support = len(datContent)/100
confidence = 0.5
```

```python
len(datContent)
```

```python
# 3rd step: 1st scan to store singletons in dictionary as candidate iteamset1 C1
C1 = {} 
for transcation in datContent:
    for each_ele in transcation:
        if each_ele not in C1:
            C1[each_ele] = 1
        else:
            C1[each_ele] += 1
```

```python
# convert key into tuple
C1_tuple = {}
for key,value in C1.items():
    C1_tuple[(key,)] = value
```

```python
# 4th step: for loop to generate Ck, Lk
frequent_itemset = {} #L
candidate_itemset = {} #C
Ck = C1_tuple
i=1
candidate_itemset[i] = C1
while 1:
    if len(Lk)==1:
        break
    Lk = generate_Lk(Ck,support)  #L1 <- C1
    frequent_itemset[i] = Lk
    Ck = generate_Ck(datContent,Lk) # C2 <- L1
    i += 1
    candidate_itemset[i] = Ck
    
```

```python
# generate association rules according to frequent_itemset
def generate_rule(frequent_itemset,confidence):
    # convert all frequent_itemset into list
    rule = {}
    all_frequent_itemset = [] # list to store all subsets in one list
    for i in frequent_itemset.values():
        for j in i.keys():
            all_frequent_itemset.append(j)
    # for loop list twice to combine subset --> generate rules
    for i in range(len(all_frequent_itemset)):
        for j in range(len(all_frequent_itemset)):
            # calculate confidence, filter confidence > threshold
            if all_frequent_itemset[i] not in all_frequent_itemset[j] and all_frequent_itemset[j] not in all_frequent_itemset[i]: # no repetation
                tuple_combine = all_frequent_itemset[i]+all_frequent_itemset[j]
                # order tuple
                temp_list_t1 = [int(i) for i in tuple_combine]
                temp_list_t1.sort()
                temp_list_t2 = [str(i) for i in temp_list_t1 ]
                tuple_combine_sorted = tuple(temp_list_t2)
                # in which k-itemset
                len1 = len(tuple_combine_sorted)  
                len2 = len(all_frequent_itemset[i])
                support1 =1
                support2 =0
                if len1<=len(frequent_itemset) and len2<=len(frequent_itemset):
                    if tuple_combine_sorted in frequent_itemset[len1].keys():
                        support1 = frequent_itemset[len1][tuple_combine_sorted]
                    if all_frequent_itemset[i] in frequent_itemset[len2].keys():
                        support2 = frequent_itemset[len2][all_frequent_itemset[i]]
                    conf = support1/support2
                    if conf>confidence:
                        rule[all_frequent_itemset[i]]=all_frequent_itemset[j]
    return rule
    
```

```python
rule = generate_rule(frequent_itemset,confidence)
```
